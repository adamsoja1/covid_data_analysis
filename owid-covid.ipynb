```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('owid-covid-data.csv')
df_filtered = df.loc[df['location'].isin(['Poland','Japan'])] 
```

```python
plt.figure(figsize=(25,15))
plot_ = sns.lineplot(data=df_filtered, x="date", y="new_deaths_smoothed_per_million", hue='location')
for ind, label in enumerate(plot_.get_xticklabels()):
    if ind % 100 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
plt.xlabel("Dates")
plt.ylabel("Deaths per milion")
plt.show()
```

```python
plt.figure(figsize=(15, 8))

# Create a bar plot with dates on the x-axis
plot_= sns.lineplot(data=df_filtered, x="date", y="stringency_index", hue='location', palette="viridis")

for ind, label in enumerate(plot_.get_xticklabels()):
    if ind % 90 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
plt.title('Stringency index over time')
plt.xlabel("Dates")
plt.ylabel("Stringency inddex")
plt.xticks(rotation=45, ha="right")
plt.show()
```
