```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('owid-covid-data.csv')
df_filtered = df.loc[df['location'].isin(['Poland','Japan'])] 
```

```python
plt.figure(figsize=(30,15))
plot_ = sns.lineplot(data=df_filtered, x="date", y="new_deaths_smoothed_per_million", hue='location')
for ind, label in enumerate(plot_.get_xticklabels()):
    if ind % 100 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
plt.show()
```

```python
plt.figure(figsize=(30,15))
plot_ = sns.lineplot(data=df_filtered, x="date", y="new_vaccinations_smoothed_per_million", hue='location')
for ind, label in enumerate(plot_.get_xticklabels()):
    if ind % 100 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
plt.show()
```
